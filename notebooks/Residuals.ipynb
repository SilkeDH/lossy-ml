# Residuals

## Quantization

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
with open('../results/output/CF_quantization.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
cf_method1 = []
cf_method2 = []

for i in [0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 1.0]:
    cf_method1.append(np.mean(data['conv_pq']['cf_' + str(i)]))
    cf_method2.append(np.mean(data['conv_me']['cf_' + str(i)]))
```

```python
plt.rcParams.update({'font.size': 17})
fig = plt.figure(figsize=(20,10))
cf = np.arange(8)
ax = fig.add_axes([0,0,1,1])
pps = ax.bar(cf - 0.15, cf_method1, width = 0.30, label = 'Method 1')
pps1 = ax.bar(cf + 0.15, cf_method2, width = 0.30, label = 'Method 2')
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1:8] = ['0.0001', '0.001', '0.01', '0.1', '0.3', '0.5', '0.7', '1']
width = 0.30
for p in pps:
    height = p.get_height()
    ax.annotate('{}'.format(round(height,1)),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')
    
    
for p in pps1:
    height = p.get_height()
    ax.annotate('{}'.format(round(height,1)),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom') 
    
ax.set_xticklabels(labels)
fig.legend(bbox_to_anchor=(0.01, 1.03, 1, -0.1), loc=2,
           ncol=1)
plt.grid()
plt.xlabel('Abs. Error (K)')
plt.ylabel('Compression Factor')
plt.title('Quantization: Method 1 vs. Method 2')
```

## Encoder

```python
with open('../results/output/CF_encoder.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
cf_bz = []
cf_zlib= []
cf_lzma = []

for i in [0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 1.0]:
    cf_bz.append(np.mean(data['bz2']['cf_' + str(i)]))
    cf_zlib.append(np.mean(data['zlib']['cf_' + str(i)]))
    cf_lzma.append(np.mean(data['lzma']['cf_' + str(i)]))
```

```python
plt.rcParams.update({'font.size': 17})
fig = plt.figure(figsize=(20,10))
cf = np.arange(8)
ax = fig.add_axes([0,0,1,1])
pps = ax.bar(cf - 0.25, cf_bz, width = 0.25, label = 'bzip2')
pps1= ax.bar(cf + 0.00, cf_zlib, width = 0.25, label = 'zlib')
pps2 = ax.bar(cf + 0.25, cf_lzma, width = 0.25, label = 'lzma')
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1:8] = ['0.0001', '0.001', '0.01', '0.1', '0.3', '0.5', '0.7', '1']
width = 0.25
for p in pps:
    height = p.get_height()
    ax.annotate('{}'.format(round(height,1)),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')
    
    
for p in pps1:
    height = p.get_height()
    ax.annotate('{}'.format(round(height,1)),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom') 
    
for p in pps2:
    height = p.get_height()
    ax.annotate('{}'.format(round(height,1)),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')
ax.set_xticklabels(labels)
fig.legend(bbox_to_anchor=(0.01, 1.03, 1, -0.1), loc=2,
           ncol=1)
plt.grid()
plt.xlabel('Abs. Error (K)')
plt.ylabel('Compression Factor')
plt.title('Bzip2 vs zlib vs lzma')
```
