```python
import pickle
import random
import matplotlib.pyplot as plt

model_europe = ['model_7', 'model_10', 'model_6', 'model_5', 'model_4']

model_globe = ['model_8', 'model_9', 'model_11', 'model_12', 'model_13']


def get_values(model_, mode):
    with open('results/'+model_+'/model-history.pkl', 'rb') as f:
        data = pickle.load(f)
    val1 = []
    val2 = []
    val3 = [] 
    for values in data['model'][mode]:
        val1.append(values['loss'])
        val2.append(values['r2_coef'])
        val3.append(values['MAE'])
    return val1, val2, val3

epochs = range(1, 201)

loss, r2, mae = get_values('model_7', 'train')
val_loss, val_r2, val_mae = get_values('model_7', 'test')


loss_2, r2_2, mae_2 = get_values('model_10', 'train')
val_loss_2, val_r2_2, val_mae_2 = get_values('model_10', 'test')


loss_3, r2_3, mae_3 = get_values('model_6', 'train')
val_loss_3, val_r2_3, val_mae_3 = get_values('model_6', 'test')


loss_4, r2_4, mae_4 = get_values('model_5', 'train')
val_loss_4, val_r2_4, val_mae_4 = get_values('model_5', 'test')

loss_5, r2_5, mae_5 = get_values('model_4', 'train')
val_loss_5, val_r2_5, val_mae_5 = get_values('model_4', 'test')

plt.figure(figsize=(10,10))
plt.plot(epochs, loss, '-r',label = "CF: 3.2 - Train")
plt.plot(epochs, val_loss, '--r', label = "CF: 3.2 - Test")

plt.plot(epochs, loss_2, '-b', label = "CF: 25.6 - Train")
plt.plot(epochs, val_loss_2, '--b', label = "CF: 25.6 - Test")

plt.plot(epochs, loss_3, '-k', label = "CF: 64 - Train")
plt.plot(epochs, val_loss_3, '--k', label = "CF: 64 - Test")

plt.plot(epochs, loss_4, '-g', label = "CF: 102.4 - Train")
plt.plot(epochs, val_loss_4, '--g', label = "CF: 102.4 - Test")

plt.plot(epochs, loss_5, '-c', label = "CF: 204.8 - Train")
plt.plot(epochs, val_loss_5, '--c', label = "CF: 204.8 - Test")

plt.xlabel('Epochs')
plt.yscale('log')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Training vs Validation loss - 5 models')
plt.legend()
plt.show()
```
